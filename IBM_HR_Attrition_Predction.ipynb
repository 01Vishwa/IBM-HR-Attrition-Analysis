```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)
from imblearn.over_sampling import SMOTE
```

```python
data=pd.read_csv("/content/train delete.csv")
```

```python
data.head()
```

```python
data = data.drop(['MonthlyRate', 'DailyRate','HourlyRate','Over18','EmployeeCount','StandardHours','EmployeeNumber'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['Attrition'] = le.fit_transform(data['Attrition'])
data['BusinessTravel'] = le.fit_transform(data['BusinessTravel'])
data['Department'] = le.fit_transform(data['Department'])
data['EducationField'] = le.fit_transform(data['EducationField'])
data['Gender'] = le.fit_transform(data['Gender'])
data['JobRole'] = le.fit_transform(data['JobRole'])
data['MaritalStatus'] = le.fit_transform(data['MaritalStatus'])
data['OverTime'] = le.fit_transform(data['OverTime'])
```

```python
features=data.drop('Attrition',axis=1)
target=data['Attrition']
```

```python
features
```

```python
target
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

target = pd.Series(target)

smote = SMOTE(random_state=42)

oversampled_features, oversampled_target = smote.fit_resample(features, target)

print("Original data shape:", features.shape, target.shape)
print("Oversampled data shape:", oversampled_features.shape, oversampled_target.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.1,random_state=42)
```

```python
from sklearn.svm import SVC
svm=SVC()
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
cr=classification_report
ac=accuracy_score
```

```python
svm.fit(X_train,y_train)
y_pred=svm.predict(X_test)
print(classification_report(y_test,y_pred))
ac(y_test,y_pred)
```

```python
svm.score(X_train,y_train)
```
